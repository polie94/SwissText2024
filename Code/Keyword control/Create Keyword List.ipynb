```python
df=pd.DataFrame(columns=["positive","negative","mixed"])
random.seed(42)
for i in range(0,50):

  sub=random.sample(words,1)

  pos=random.sample(positive_adjektive,2)
  pos.extend(sub)
  df.loc[i,"positive"]=pos
  neg=random.sample(negative_adjektive,2)
  neg.extend(sub)
  df.loc[i,"negative"]=neg
  mix=[]
  mix.extend(random.sample(positive_adjektive,1))
  mix.extend(random.sample(negative_adjektive,1))
  mix.extend(sub)
  df.loc[i,"mixed"]=mix
df.to_csv("/content/drive/MyDrive/Masterthesis/Data/sent_keywords.tsv", index=False)
```
